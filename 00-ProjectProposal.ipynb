## Research Question

Instructions: REPLACE the contents of this cell with your research question

What extent do NBA Draft Combine athletic measurements predict players' first five years of career performance, after controlling for draft position and college statistics?


## Background and Prior Work

The NBA Draft Combine is a key evaluation tool for teams to assess prospective players' physical and skill-based attributes by using standardized measurements and athletic tests, like height, wingspan, vertical jump, three-quarter-court sprint, and on-court drills. (1) The draft combine results are often used to evaluate a player’s physical potential and readiness for the NBA besides the college statistics. 

Certain physical traits may provide advantages that are not fully reflected in college performance data, but it is unclear to what extent or which draft combined measurements could translate into sustained NBA success. Recent studies about NBA Draft Combine metrics have found weak or non-significant relationships between athletic test results and rookie performance indicators, including points, rebounds, and assists per game. This finding suggests that while combined data may capture athletic potential, it fails to account for intangible factors such as basketball intelligence, adaptability, and in-game decision-making, limiting its ability to predict early professional performance.

In addition, other prior research has shown that draft position itself is strongly associated with long-term career outcomes; particularly the top picks tend to have significantly longer NBA careers, with career longevity decreasing as draft standing increases. (2) This shows that draft position captures teams’ overall evaluations of talent and potential and should be controlled when assessing the predictive value of pre-draft combine measurements.

After controlling for draft position and college performance, this project explores whether pre-draft combine measurements can predict a player’s on-court performance during the first five years of their NBA career.

References:
1.https://www.researchgate.net/publication/396582065_Evaluating_the_Predictive_Power_of_NBA_Draft_Combine_Metrics_on_Early-Career_Performance_An_Empirical_Analysis_from_the_20222024_Seasons

2.https://archives.rpd-online.com/article/download/v28-n3-miguel-milan-soares-etal/2746-14000-1-PB.pdf

## Hypothesis


We predict that after controlling for draft position and college statistics, NBA Draft Combine athletic measurements will have limited ability to predict players performance during their first five years in the league. Prior research shows that draft position is strongly linked to career outcomes (Reference 1), while combine metrics have weak and often non significant relationships with early career performance (Reference 2). Therefore, we expect that most combined measurements will explain little additional variation in long term performance beyond what is already captured by draft order and college success.
